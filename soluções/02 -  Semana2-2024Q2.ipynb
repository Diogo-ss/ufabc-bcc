```
import numpy as np
import matplotlib.pyplot as plt
import math

orange = '#ea9b98'

## modo claro
# pink = '#dd7878'

# base = '#eff1f5'
# text = '#4c4f69'
# maroon = '#e64553'
# peach = '#fe640b'
# mauve = '#8839ef'

# modo escuro
bg = '#222035'
fg = '#8b8aa1'

pink = "#ec6f92"

base = '#1e1e2e'
text = '#cad3f5'
maroon = '#ee99a0'
peach	= '#f5a97f'
mauve	= '#c6a0f6'
```

1) $y=x^2$  
1.1) $y2=x^3$


```
x= np.linspace(-10,10,50)
y= np.power(x,2)
y2= np.power(x,3)
# Acabei de criar as coordenadas
plt.figure(dpi=130, facecolor = 'y')
plt.plot(x,y,'k-',label= '$y=x^2$',linewidth= 2)
plt.plot(x,y2,'r-',label= '$y2=x^3$',linewidth= 3)
plt.xticks(rotation=85)
plt.yticks(rotation=35)
plt.legend()
plt.grid(color='r', linestyle='-.', linewidth=1)
```

```
x = np.linspace(-10, 10, 100)

fig, ax = plt.subplots(figsize=(16, 9), facecolor=base)

ax.set_facecolor(base)
ax.tick_params(which='both', colors=text)

for axis in ['top', 'bottom', 'left', 'right']:
  ax.spines[axis].set_color(fg)

ax.plot(x, x**2, label='Quadrática ($f(x)=x^2$)', color=orange)
ax.plot(x, x**3, label='Cúbica ($f\'(x)=x^3$)', color=pink)
ax.set_xlabel('X', color=text)
ax.set_ylabel('Y', color=text)

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color=fg)

legend = ax.legend(facecolor=base, labelcolor=text, edgecolor=fg)
legend.get_frame().set_alpha(1)


plt.show()
```

2) $y=sin(x)$


```
x= np.linspace(0, 2* np.pi ,40)
y= np.sin(x)

plt.figure()
plt.plot(x,y,'bh',label='$y=sin(x)$',linewidth= 1.1)
plt.legend()
```

```
x = np.arange(-4 * math.pi, 4 * math.pi,0.1)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(16, 9), facecolor=base)

ax.set_facecolor(base)
ax.tick_params(which='both', colors=text)

ax.axhline(0, color=fg, linewidth=1.7)
ax.axvline(0, color=fg, linewidth=1.7)

for axis in ['top', 'bottom', 'left', 'right']:
  ax.spines[axis].set_color(fg)

ax.plot(x, y, label='Seno ($f(x)=sin(x)$)', color=maroon)

ax.set_ylim(-1.1, 1.1)

ax.set_title('Gráfico da Função Seno', color=text)

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color=fg)

legend = ax.legend(facecolor=base, labelcolor=text, edgecolor=fg)
legend.get_frame().set_alpha(1)

plt.show()
```

3) $y=x^2 -5x +6$


```
x= np.linspace(-100,100,1000)
y= np.power(x,2) - np.multiply(x,5) + 6

plt.figure()
plt.plot(x,y,'g-',label='$y=x^2 -5x +6$',linewidth= 1.2)
plt.legend()
```

```
x = np.linspace(-100, 100, 1000)
y= x**2 - np.multiply(x,5) + 6

fig, ax = plt.subplots(figsize=(16, 9), facecolor=base)

ax.set_facecolor(base)
ax.tick_params(which='both', colors=text)

for axis in ['top', 'bottom', 'left', 'right']:
  ax.spines[axis].set_color(fg)

ax.plot(x, y, label='$f(x)=x^2-5x+6$', color=orange)
ax.set_xlabel('X', color=text)
ax.set_ylabel('Y', color=text)

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color=fg)

legend = ax.legend(facecolor=base, labelcolor=text, edgecolor=fg)
legend.get_frame().set_alpha(1)


plt.show()
```

4) $f_1(x)=\frac{x}{1+x^2}$




```
x= np.linspace(0,100,1000)
y= x/(1 + np.power(x,2))

plt.figure()
plt.plot(x,y,'c-',label='$f_1(x)$',linewidth= 1.0)
plt.legend(loc='best')
```

onde $f_1(x)= \frac{x}{1+x^2}$ representa a funcao fualana que estou desenhando...


```
x = np.linspace(0, 100, 1000)
y= x/(1 + x**2)

fig, ax = plt.subplots(figsize=(16, 9), facecolor=base)

ax.set_facecolor(base)
ax.tick_params(which='both', colors=text)

for axis in ['top', 'bottom', 'left', 'right']:
  ax.spines[axis].set_color(fg)

ax.plot(x, y, label='$f_1(x)=\\frac{x}{1+x^2}$', color=orange)
ax.set_xlabel('X', color=text)
ax.set_ylabel('Y', color=text)

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color=fg)

legend = ax.legend(facecolor=base, labelcolor=text, edgecolor=fg)
legend.get_frame().set_alpha(1)


plt.show()
```
